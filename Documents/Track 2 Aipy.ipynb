```python
# FAST CIFAR-10 Object Recognition - Optimized for Speed
# Track 2: Deep Learning (CNNs & Transfer Learning)

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import pandas as pd
import time

print("TensorFlow version:", tf.__version__)

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# =============================================================================
# 1. FAST DATA PREPARATION
# =============================================================================

def load_data_fast():
    """Load and quickly preprocess data"""
    print("Loading CIFAR-10 dataset...")
    (x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

    # Normalize pixel values
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0

    # Convert labels to categorical
    y_train_cat = keras.utils.to_categorical(y_train, 10)
    y_test_cat = keras.utils.to_categorical(y_test, 10)

    # Use smaller subset for faster training (comment out for full training)
    x_train = x_train[:10000]  # Use only 10K samples instead of 50K
    y_train_cat = y_train_cat[:10000]

    # Split training data
    x_train_split, x_val, y_train_split, y_val = train_test_split(
        x_train, y_train_cat, test_size=0.2, random_state=42
    )

    class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
                   'dog', 'frog', 'horse', 'ship', 'truck']

    print(f"Training samples: {x_train_split.shape[0]}")
    print(f"Validation samples: {x_val.shape[0]}")
    print(f"Test samples: {x_test.shape[0]}")

    return (x_train_split, y_train_split), (x_val, y_val), (x_test, y_test_cat), class_names

# =============================================================================
# 2. FAST CUSTOM CNN (Lightweight)
# =============================================================================

def create_fast_cnn():
    """Create a lightweight CNN for fast training"""
    model = keras.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# =============================================================================
# 3. FAST TRANSFER LEARNING (Using smaller pre-trained model)
# =============================================================================

def create_fast_transfer_learning():
    """Create transfer learning model with MobileNetV2 (faster than VGG16)"""
    # Use MobileNetV2 which is much faster
    base_model = keras.applications.MobileNetV2(
        weights='imagenet',
        include_top=False,
        input_shape=(32, 32, 3)  # Using original 32x32 for speed
    )

    # Freeze base model
    base_model.trainable = False

    model = keras.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model, base_model

# =============================================================================
# 4. FAST TRAINING FUNCTIONS
# =============================================================================

def train_fast_cnn(model, x_train, y_train, x_val, y_val):
    """Fast training for CNN"""
    print("Training Custom CNN (Fast)...")
    start_time = time.time()

    history = model.fit(
        x_train, y_train,
        batch_size=64,  # Smaller batch for speed
        epochs=10,       # Only 10 epochs
        validation_data=(x_val, y_val),
        verbose=1
    )

    training_time = time.time() - start_time
    print(f"Custom CNN training completed in {training_time:.2f} seconds")
    return history

def train_fast_transfer_learning(model, base_model, x_train, y_train, x_val, y_val):
    """Fast training for transfer learning"""
    print("Training Transfer Learning Model (Fast)...")
    start_time = time.time()

    # Phase 1: Train top layers
    history_phase1 = model.fit(
        x_train, y_train,
        batch_size=64,
        epochs=5,  # Only 5 epochs for phase 1
        validation_data=(x_val, y_val),
        verbose=1
    )

    # Phase 2: Quick fine-tuning (optional - skip for even faster execution)
    print("Skipping fine-tuning for maximum speed...")

    training_time = time.time() - start_time
    print(f"Transfer Learning training completed in {training_time:.2f} seconds")

    return history_phase1, None

# =============================================================================
# 5. QUICK VISUALIZATION
# =============================================================================

def quick_visualize_predictions(model, x_test, y_test, class_names, model_name):
    """Quick visualization of predictions"""
    # Get predictions for first 10 samples
    predictions = model.predict(x_test[:10])
    pred_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(y_test[:10], axis=1)

    plt.figure(figsize=(12, 4))
    for i in range(10):
        plt.subplot(2, 5, i + 1)
        plt.imshow(x_test[i])

        true_class = class_names[true_classes[i]]
        pred_class = class_names[pred_classes[i]]
        confidence = np.max(predictions[i])

        color = 'green' if true_class == pred_class else 'red'
        plt.title(f'True: {true_class}\nPred: {pred_class}', color=color, fontsize=8)
        plt.axis('off')

    plt.suptitle(f'{model_name} - Sample Predictions', fontweight='bold')
    plt.tight_layout()
    plt.show()

def plot_quick_comparison(cnn_accuracy, tl_accuracy):
    """Quick bar chart comparison"""
    models = ['Custom CNN', 'Transfer Learning']
    accuracies = [cnn_accuracy, tl_accuracy]

    plt.figure(figsize=(8, 5))
    bars = plt.bar(models, accuracies, color=['lightblue', 'lightcoral'])
    plt.ylim(0, 1)
    plt.ylabel('Test Accuracy')
    plt.title('Model Performance Comparison (Fast Training)')

    # Add value labels on bars
    for bar, acc in zip(bars, accuracies):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{acc:.3f}', ha='center', va='bottom', fontweight='bold')

    plt.show()

# =============================================================================
# 6. MAIN FAST EXECUTION
# =============================================================================

def main_fast():
    """Main function optimized for speed"""
    print("FAST CIFAR-10 OBJECT RECOGNITION")
    print("=" * 50)

    # Load data quickly
    (x_train, y_train), (x_val, y_val), (x_test, y_test), class_names = load_data_fast()

    # Create and train Custom CNN
    print("\n1. CREATING FAST CUSTOM CNN")
    custom_cnn = create_fast_cnn()
    custom_cnn.summary()

    history_cnn = train_fast_cnn(custom_cnn, x_train, y_train, x_val, y_val)

    # Create and train Transfer Learning model
    print("\n2. CREATING FAST TRANSFER LEARNING MODEL")
    tl_model, base_model = create_fast_transfer_learning()
    tl_model.summary()

    history_tl, _ = train_fast_transfer_learning(tl_model, base_model, x_train, y_train, x_val, y_val)

    # Quick evaluation
    print("\n3. QUICK EVALUATION")

    # Evaluate Custom CNN
    cnn_test_loss, cnn_test_acc = custom_cnn.evaluate(x_test, y_test, verbose=0)
    print(f"Custom CNN Test Accuracy: {cnn_test_acc:.4f}")

    # Evaluate Transfer Learning
    tl_test_loss, tl_test_acc = tl_model.evaluate(x_test, y_test, verbose=0)
    print(f"Transfer Learning Test Accuracy: {tl_test_acc:.4f}")

    # Quick visualizations
    print("\n4. GENERATING QUICK VISUALIZATIONS")
    quick_visualize_predictions(custom_cnn, x_test, y_test, class_names, "Custom CNN")
    quick_visualize_predictions(tl_model, x_test, y_test, class_names, "Transfer Learning")

    # Performance comparison
    plot_quick_comparison(cnn_test_acc, tl_test_acc)

    # Generate predictions for analysis
    cnn_pred = custom_cnn.predict(x_test)
    tl_pred = tl_model.predict(x_test)

    cnn_pred_classes = np.argmax(cnn_pred, axis=1)
    tl_pred_classes = np.argmax(tl_pred, axis=1)
    true_classes = np.argmax(y_test, axis=1)

    # Quick classification reports
    print("\nCUSTOM CNN - Quick Classification Report:")
    print(classification_report(true_classes, cnn_pred_classes,
                              target_names=class_names, digits=3))

    print("\nTRANSFER LEARNING - Quick Classification Report:")
    print(classification_report(true_classes, tl_pred_classes,
                              target_names=class_names, digits=3))

    return custom_cnn, tl_model, cnn_test_acc, tl_test_acc

# =============================================================================
# 7. ULTRA-FAST DEMONSTRATION (For instant results)
# =============================================================================

def ultra_fast_demo():
    """Ultra-fast demonstration using pre-trained models if available"""
    print("ULTRA-FAST DEMONSTRATION")
    print("=" * 40)

    # Load data
    (x_train, y_train), (x_val, y_val), (x_test, y_test), class_names = load_data_fast()

    # Option 1: Use even smaller model for instant results
    print("Creating ultra-fast model...")
    ultra_model = keras.Sequential([
        layers.Conv2D(16, (3, 3), activation='relu', input_shape=(32, 32, 3)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(32, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])

    ultra_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Train for just 3 epochs
    print("Training for 3 epochs (ultra-fast)...")
    ultra_model.fit(x_train, y_train, epochs=3, batch_size=32, validation_data=(x_val, y_val), verbose=1)

    # Quick evaluation
    test_loss, test_acc = ultra_model.evaluate(x_test, y_test, verbose=0)
    print(f"Ultra-fast model test accuracy: {test_acc:.4f}")

    # Show some predictions
    print("\nSample predictions:")
    sample_indices = [0, 1, 2, 3, 4]
    for idx in sample_indices:
        prediction = ultra_model.predict(np.expand_dims(x_test[idx], axis=0))[0]
        pred_class = class_names[np.argmax(prediction)]
        true_class = class_names[np.argmax(y_test[idx])]
        confidence = np.max(prediction)

        correct = "✓" if pred_class == true_class else "✗"
        print(f"  Sample {idx}: True={true_class}, Pred={pred_class} {correct} (conf: {confidence:.2f})")

    return ultra_model, test_acc

# =============================================================================
# 8. COMPARATIVE ANALYSIS
# =============================================================================

def quick_analysis(cnn_acc, tl_acc):
    """Quick comparative analysis"""
    print("\n" + "=" * 50)
    print("QUICK COMPARATIVE ANALYSIS")
    print("=" * 50)

    improvement = tl_acc - cnn_acc
    improvement_percent = (improvement / cnn_acc) * 100

    print(f"Custom CNN Accuracy: {cnn_acc:.4f}")
    print(f"Transfer Learning Accuracy: {tl_acc:.4f}")
    print(f"Improvement: {improvement:.4f} ({improvement_percent:.1f}%)")

    print("\nKEY INSIGHTS:")
    if improvement > 0.05:
        print("✓ Transfer Learning provides significant improvement")
    elif improvement > 0.02:
        print("✓ Transfer Learning provides moderate improvement")
    else:
        print("∼ Transfer Learning provides minimal improvement in this quick training")

    print("✓ Both approaches demonstrate the power of deep learning")
    print("✓ Transfer learning leverages pre-trained features effectively")
    print("✓ Custom CNN trains faster but may have lower accuracy")

# =============================================================================
# EXECUTION OPTIONS
# =============================================================================

if __name__ == "__main__":
    print("Choose execution mode:")
    print("1. Fast Training (Recommended - 2-5 minutes)")
    print("2. Ultra-Fast Demo (Instant results - 30-60 seconds)")

    choice = input("Enter choice (1 or 2): ").strip()

    if choice == "2":
        # Ultra-fast demonstration
        model, accuracy = ultra_fast_demo()
        print(f"\nUltra-fast demo completed! Final accuracy: {accuracy:.4f}")
    else:
        # Fast training (default)
        cnn_model, tl_model, cnn_acc, tl_acc = main_fast()
        quick_analysis(cnn_acc, tl_acc)

        # Save models for later use
        cnn_model.save('fast_cnn_model.h5')
        tl_model.save('fast_tl_model.h5')
        print("\nModels saved as 'fast_cnn_model.h5' and 'fast_tl_model.h5'")

    print("\n" + "=" * 50)
    print("EXECUTION COMPLETED!")
    print("=" * 50)
```
